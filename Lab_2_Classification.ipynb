```python
import numpy as np
import pandas as pd
```

```python
#(1) "gapminder.csv" 파일로부터 gapminder 변수에 데이터 로드
gapminder = pd.read_csv('gapminder.csv')
display(gapminder)
```

```python
#(2) 데이터 타입확인
print(gapminder.dtypes)
```

```python
#(3)데이터 구조확인 (행, 열 레이블, 행,열 개수)
print("행 레이블 = \n", gapminder.index) #행 레이블
print("열 레이블 = \n", gapminder.columns) #열 레이블
print("행, 열 갯수 = \n", gapminder.shape) #행, 열 개수
```

```python
#(4) 데이터 타입, 구조, 누락값 동시에 확인
gapminder.info()
```

```python
#(5)누락된 값 있는 행삭제
df = gapminder.isnull()
print(df)
print(gapminder.dropna())
```

```python
#(6)데이터 내용 일부 확인 (상위 12개, 하위 5개)
display("#gapminder.head=", gapminder.head(12))
display("#gapminder.tail=", gapminder.tail(5))
```

```python
#(7)요약통계
gapminder.describe()
```

```python
#(8) 2007년도 데이터만 추출
df = gapminder[gapminder['year'] == 2007]
display(df)
```

```python
#(9)데이터 일부 추출, 한국에 대한 2002,2007년 데이터만 추출
df = gapminder[(gapminder['country'] == 'Korea Rep.') & ((gapminder['year'] == 2007) | (gapminder['year'] == 2002))]
display(df)
```

```python
#(10)국가 'country', 년도 'year', 인구수 'pop' 만 추출
df = gapminder[['country', 'year', 'pop']]
display(df)
```

```python
#(11)년도순으로 데이터 정렬
df = gapminder.sort_values('year')
display(df)
```

```python
#(12)인구수 역순으로 데이터 정렬
df = gapminder.sort_values('pop', ascending=True)
display(df)
```

```python
#(13)년도, 대륙순으로 데이터 정렬
df = gapminder.sort_values(['year', 'continent'])
display(df)
```

```python
#(14)대륙별로 그룹처리하여 그룹에 대한 중앙값, 평균, 표준편차 요약통계
grouped = gapminder.groupby('continent')

for continent, group in grouped:
    print("groupname: ", continent)
    print("groupdata: ")
    print(group)

v1 = grouped.get_group(name= 'Asia')
v2 = grouped.get_group(name= 'Americas')
v3 = grouped.get_group(name = 'Europe')
v4 = grouped.get_group(name = 'Oceania')
v5 = grouped.get_group(name = 'Africa')

print(grouped.median())
print(grouped.mean())
print(grouped.std())
```

```python
#(15)국가별, 년도에 대한 GDP 테이블로 재구성
pf = pd.pivot_table(gapminder, index='country', columns= 'year', values= 'gdpPercap')
display("국가별, 년도에 대한 GDP 테이블", pf)
```

```python
#(16)년도별 대륙에 대한 인구수 평균 테이블로 재구성
pf = pd.pivot_table(gapminder, index='year', columns= 'continent', values= 'pop', aggfunc = np.max)
display("년도별 대륙에 대한 인구수 평균 테이블", pf)
```

```python
#(17)continent(대륙)을 범주형으로 변환
gapminder['continent'] = gapminder['continent'].astype("category")
print(gapminder.dtypes)
display(gapminder)
```

```python
#(18)년도별 대륙에 대한 GDP 평균 테이블로 재구성
pf = pd.pivot_table(gapminder, index='year', columns= 'continent', values= 'gdpPercap', aggfunc = np.mean)
display("년도별 대륙에 대한 GDP 평균 테이블", pf)
```
