### JSON (JavaScript Object Notation)
 
 - 자바스크립트 클라이언트와 백엔드 서버 간의 통신을 위한 형식으로 개발   
 - 텍스트 기반 데이터를 기술하고 교환하는 가장 보편적인 매체  
 - 로그인정보를 표현한 json  
  
 

- 중첩된 형태의 json


```python
import json

#딕셔너리 자료형
d = {"name":"홍길동", "birth":"0525", "age": 30}
print("d=",d)

#json.dumps() : 딕셔너리 자료형을 JSON 문자열로 생성 
json_data = json.dumps(d)
print(json_data)

# json.loads(): json 데이터를 딕셔너리 자료형으로 반환
dd = json.loads(json_data)
print("dd=",dd)

# ensure_ascii=False 옵션을 사용하여 데이터를 저장할 때 아스키 형태로 변환하지 않고 저장
json_data2 = json.dumps(d, ensure_ascii=False)
print(json_data2)
# json.loads(): json 데이터를 딕셔너리 자료형으로 반환
dd = json.loads(json_data2)
print("dd=",dd)
```

### 동물등록 현황 openAPI 활용 코드

```python
## 필요한 라이브러리 임포트
import requests
import json
import pandas as pd
from  pandas import json_normalize
```

#### 1) 10개만 요청하여 확인

```python
#url 입력, 개인인증키에는 본인이 부여받은 인증키를 입력
#5개 데이터만 일부요청 확인
url = "http://211.237.50.150:7080/openapi/da28e3097b56ba49ea7e17aee58fa8582e12dd96906465ee56c1d087abeec73f/json/Grid_20210806000000000612_1/1/10"
print (url)

#url 불러오기
response = requests.get(url)

#데이터값 가져오기
contents = response.text

##  데이터 출력
print("contents =", contents)
print("contents type = ", type(contents))
print()

#읽은 데이터 변환, json형의 contents를 딕셔너리 자료형으로 반환
rdata = json.loads(contents)
print("rdata = ", rdata)
print("rdata type = ", type(rdata))

t = rdata['Grid_20210806000000000612_1']['totalCnt']
print('totalCnt = ', t)
```

```python

#필요한 데이터만 접근하여 추출
#Grid_20210806000000000612_1'의 row 키접근
d = rdata['Grid_20210806000000000612_1']['row']
#print(d)

#접근한 데이터를 데이터프레임으로 추출
data = json_normalize(d)

#데이터 구조확인
print("data.shape=", data.shape)            #행, 열의 개수 
print("data.index=",data.index)             #행 레이블
print("data.columns=", data.columns)        #열 레이블 
print("data.dtypes=", data.dtypes)          #칼럼(속성) 타입 

#데이터 행, 열 정보 확인
print("data.info()=", data.info())   

#데이터 내용 일부  확인
print(data.head(5))
print(data.tail(5))


## dataframe으로 저장된 데이터 csv로 저장
#df.to_csv('동물등록현황_202303.csv',index=False)
```

#### 2) 1000개씩 요청하여 결합

```python
#1000개 단위로 요청
it = t // 1000
r = t % 1000
print (it, r)
```

```python
#url 입력, 개인인증키에는 본인이 부여받은 인증키를 입력
data = pd.DataFrame()
add = "http://211.237.50.150:7080/openapi/da28e3097b56ba49ea7e17aee58fa8582e12dd96906465ee56c1d087abeec73f/json/Grid_20210806000000000612_1/"
#1000개씩 데이터 요청 반복
for i in range(it):
    st = i * 1000 + 1
    end = (i+1) * 1000
    url = f'{add}{st}/{end}'
    response = requests.get(url)
    contents = response.text
    #읽은 데이터 변환, json형의 contents를 딕셔너리 자료형으로 반환
    rdata = json.loads(contents)
    #필요한 데이터만 접근하여 추출
    #Grid_20210806000000000612_1'의 row 키접근
    d = rdata['Grid_20210806000000000612_1']['row']
    #접근한 데이터를 데이터프레임으로 추출
    data1 = json_normalize(d)
    data = data.append(data1, ignore_index=True)
```

```python
    
#데이터 구조확인
print("data.shape=", data.shape)            #행, 열의 개수 
print("data.index=",data.index)             #행 레이블
print("data.columns=", data.columns)        #열 레이블 
print("data.dtypes=", data.dtypes)          #칼럼(속성) 타입 

#데이터 행, 열 정보 확인
print("data.info()=", data.info())   

#데이터 내용 일부  확인
print(data.head())
print(data.tail())
```

```python
#나머지 데이터 요청
ste = end + 1
ende = end + r
print(ste, " ", ende)
url = f'{add}{ste}/{ende}'
response = requests.get(url)
contents = response.text
#읽은 데이터 변환, json형의 contents를 딕셔너리 자료형으로 반환
rdata = json.loads(contents)
#필요한 데이터만 접근하여 추출
#Grid_20210806000000000612_1'의 row 키접근
d = rdata['Grid_20210806000000000612_1']['row']
#접근한 데이터를 데이터프레임으로 추출
data1 = json_normalize(d)
data = data.append(data1, ignore_index=True)
    

#데이터 구조확인
print("data.shape=", data.shape)            #행, 열의 개수 
print("data.index=",data.index)             #행 레이블
print("data.columns=", data.columns)        #열 레이블 
print("data.dtypes=", data.dtypes)          #칼럼(속성) 타입 

#데이터 행, 열 정보 확인
print("data.info()=", data.info())   

#데이터 내용 일부  확인
print(data.head(10))
print(data.tail(10))
```
