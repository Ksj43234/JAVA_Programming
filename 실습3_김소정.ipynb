```python
#1.(1)초등학교 교원수(시도/시/군/구), 초등학교 학생수(시도/시/군/구) 두개의 엑셀파일을 읽어 테이블에 저장
data1 = pd.read_csv('Student.csv', index_col =0)
data2 = pd.read_csv('Teacher.csv', index_col =0)

display(data1)
display(data2)
```

```python
#2.(1)에서 얻은 두 테이블로 2022년 학생수, 교원수 테이블 생성
name_list = data1['행정구역별']
table1 = data1['2022']
table2 = data2['2022']

table = pd.concat([table1, table2] ,axis =1)
table.index = name_list
display(table)
```

```python
#3. 교원 1인당 학생수 속성 추가 (학생수 / 교원수)
table['S/T'] = table['Student']/table['Teacher']
display(table)
```

```python
#4. 교원1인당 학생수 역순으로 정렬
df = table.sort_values('S/T', ascending = True)
display(df)
```
